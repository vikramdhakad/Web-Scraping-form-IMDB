```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250'
req = requests.get(url)
```

```python
Web_name = []
Year = []
Rating = []
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
names = soup.find_all("td", class_="titleColumn")
print(names)
```

```python
for i in names:
    Web_name.append(i.a.text)
    
Web_name
```

```python
for y in names:
    Year.append(y.span.text)
    
Year
```

```python
rate = soup.find_all("td", class_='ratingColumn imdbRating')
rate
for r in rate:
    Rating.append(r.strong.text)
    
Rating
```

```python
data = pd.DataFrame()
data['Name'] = Web_name
data['Year'] = Year
data['Rating'] = Rating
data
```

```python
data.to_csv('IMDB Top 250 Web Series.csv', index=False)
```
