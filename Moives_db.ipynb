```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
req = requests.get(url)
```

```python
Movie_name = []
Year = []
Rating = []
```

```python
soup = BeautifulSoup(req.content, "html.parser")
#print(soup)
```

```python
name = soup.find_all("td", class_="titleColumn")
print(name)
```

```python
for i in name:
    Movie_name.append(i.a.text)
    
Movie_name
```

```python
for y in name:
    Year.append(y.span.text)
    
Year
```

```python
rate = soup.find_all("td", class_='ratingColumn imdbRating')
rate
for r in rate:
    Rating.append(r.strong.text)
    
Rating
```

```python
data = pd.DataFrame()
data['Name'] = Movie_name
data['Year'] = Year
data['Rating'] = Rating
data
```

```python
data.to_csv('IMDB Top 250 Movies.csv', index=False)
```
